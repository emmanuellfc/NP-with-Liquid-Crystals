# Data Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
volume = []
std_volume = []
temperature = []

for i in range(8):
    temp = 6.0 + 0.2*i
    temperature.append(temp)
    press = 1.8
    file = "T_" + str(temp) + "_P_" + str(press) + "_equilibrium.log"
    data = np.genfromtxt(fname = file, skip_header=True);
    
    mean_vol = np.mean(data[:,11])
    std_vol = np.std(data[:,11])
    
    volume.append(mean_vol)
    std_volume.append(std_vol)
```

```python
plt.figure(figsize=(12,6))

plt.plot(temperature, volume)
plt.ylabel('Volume')
plt.xlabel('Temperature')
plt.title('Variación de la Presión')
plt.grid()

plt.show()
```
