# Thermodynamical Analysis

## Temperature Ramp

```python
#-----Import packages.

import numpy
from matplotlib import pyplot
```

```python
#-----Define variables

t_fin = 8.0; 
press = 1.8;
```

```python
#-----Declare the file from wich we'll extract the configuration of the system.

ramp_file = "T_" + str(t_fin) + "_P_" + str(press) + "_ramp.log"
```

```python
#-----Declare the file on wich compute all of the thermodynamical properties.

data = numpy.genfromtxt(fname = ramp_file, skip_header=True);
```

```python
#-----Compute presure, temperature and its standard deviations.

mean_p = numpy.mean(data[:,10]);
std_p = numpy.std(data[:,10]);
mean_temp = numpy.mean(data[:,9])
std_temp = numpy.std(data[:,9])
```

## Values of Pressure and Temperature

```python
print("P =", mean_p, "+-",std_p)
```

```python
print("T =", mean_temp, "+-", std_temp)
```

## Graphics

```python
pyplot.figure(figsize=(4,2.2), dpi=140);
pyplot.plot(data[:,0], data[:,10]);
pyplot.xlabel('time step');
pyplot.ylabel('Pressure');
pyplot.suptitle('')
```

```python
pyplot.figure(figsize=(4,2.2), dpi=140);
pyplot.plot(data[:,0], data[:,9]);
pyplot.xlabel('Time step');
pyplot.ylabel('Temperature');
pyplot.suptitle('')
```
